```python
from qiskit.circuit.library import QFT
from qiskit.circuit.library import CU1Gate
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer
from math import pi
```

```python
def myQFT(nqbits):
    # Crée un circuit quantique pour effectuer la transformation quantique de Fourier (QFT)
    myQFT_circuit = QuantumCircuit(nqbits)

    # Applique une porte de Hadamard (H) à chaque qubit du circuit
    for i in range(nqbits):
        myQFT_circuit.h(i)
        # Applique des portes de rotation contrôlées (CU1) aux qubits suivants
        for j in range(i + 1, nqbits):
            # Utilise une CU1Gate avec un angle de pi / (2**(j-i))
            myQFT_circuit.append(CU1Gate(pi / (2**(j-i))), [j, i])

    # Renvoie le circuit QFT créé
    return myQFT_circuit
```

```python
circuit_drawer(myQFT(4), output='mpl')
```
